## 超幾何分布


### 概要

- $N$ 個の玉が入った箱（内訳： 赤 $M$ 個、白 $N-M$ 個）から、 $n$ 個抽出する。赤を成功として、成功数 $X$　が従う分布 $HG(M,N,x)$
- 非復元無作為抽出の場合を考える。（復元無作為抽出の場合は二項分布 $Bin(n, M/N)）

### 確率質量関数

$$
  f(x) = \frac{{}_M \mathrm{C}_x \cdot {}_{N-M} \mathrm{C}_{n-x}}{{}_{N} \mathrm{C}_{n}}
$$


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import hypergeom


# 確率質量関数
# x: 試行回数
# N: 母集団
# M: 成功集団
# n: 抽出数
def probability_density(x, N, M, n):
    return hypergeom.pmf(x, N, M, n)


def show_graph(N, M, n):
    x = np.arange(1, 100, 1)
    y = probability_density(x, N, M, n)
    plt.xlabel("x", fontsize=12)
    plt.ylabel("f", fontsize=12)
    plt.ylim(0)
    plt.plot(x, y, label="N:{N}, M:{M}, n:{n}".format(N=N, M=M, n=n))
    plt.legend()


show_graph(100, 40, 50)
show_graph(400, 80, 200)
```

### 平均、分散

$$
  \begin{align}
  E[X] &= n \cdot \frac{M}{N} \\
  V[X] &= n \cdot \frac{M}{N} \cdot \left(1-\frac{M}{N}\right) \cdot \frac{N-n}{N-1}\\
  \end{align}
$$

無限母集団の場合($N \rightarrow \infty$)、二項分布における平均と分散に等しくなる。  
有限母集団の場合、有限母集団修正 $\frac{N-n}{N-1} \le 1$ により二項分布における分散よりも小さい値が得られる。
