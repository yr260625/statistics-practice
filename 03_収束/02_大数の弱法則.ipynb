## 大数の弱法則

### 概要

確率変数の列 $\{X_n\}$ が独立な同一の分布に従い、平均と分散がそれぞれ $\mu, \sigma^2$ とする。  
このとき、標本平均 $\bar X = \sum_{i=1}^{n} X_i/n$ の平均と分散の $n \to \infty$ の極限は以下のように収束する。

$$
  \begin{align}
  E[\bar X] &\to \mu \\
  V[\bar X] &= \frac{\sigma^2}{n} \to 0
  \end{align}
$$


### 例）サイコロの出目の期待値


```python
import numpy as np
import matplotlib.pyplot as plt


# サイコロの出目を返却
# n: 試行回数
def throw_dice(n):
    rng = np.random.default_rng()
    return rng.integers(1, 6, size=n, endpoint=True)


# サイコロの出目一覧の標本平均を算出
# n: 試行回数
def mean_sample(n):
    dice_results = throw_dice(n)
    return np.mean(dice_results)


# グラフ表示
def show_graph():
    x = range(0, 1000)
    y = []
    for n in x:
        y.append(mean_sample(n))
    plt.plot(x, y)
    plt.ylim(1, 6)
    plt.xlabel("n", fontsize=12)
    plt.ylabel("mean sample", fontsize=12)
    plt.axhline(3.5, color="red", linestyle="dashed")


show_graph()
```
